```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud
```

```python
#load dataset
df = pd.read_csv("/content/amazon_reviews.csv")
```

```python
#vader analyazer
sid = SentimentIntensityAnalyzer()
```

```python
df['Sentiment'] = df['reviewText'].apply(lambda x:sid.polarity_scores(x)['compound'])
```

```python
df['reviewText'] = df['reviewText'].astype(str)
```

```python
# Visualize sentiment distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Sentiment'], bins=30, color='skyblue', edgecolor='black')
plt.title('Sentiment Distribution of Amazon Reviews')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
positive_reviews = df[df['Sentiment'] > 0]['reviewText']
positive_text = ' '.join(positive_reviews)
wordcloud_positive = WordCloud(width=800, height=400, background_color='white').generate(positive_text)
```

```python
# Plot word cloud for positive reviews
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_positive, interpolation='bilinear')
plt.title('Word Cloud for Positive Amazon Reviews')
plt.axis('off')
plt.show()
```

```python
# Generate word cloud for negative reviews
negative_reviews = df[df['Sentiment'] < 0]['reviewText']
negative_text = ' '.join(negative_reviews)
wordcloud_negative = WordCloud(width=800, height=400, background_color='black', colormap='Reds').generate(negative_text)
```

```python

# Plot word cloud for negative reviews
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_negative, interpolation='bilinear')
plt.title('Word Cloud for Negative Amazon Reviews')
plt.axis('off')
plt.show()
```
